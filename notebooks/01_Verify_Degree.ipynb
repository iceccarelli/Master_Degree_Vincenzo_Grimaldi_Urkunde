I completed my Master of Science at RWTH Aachen University on **08 August 2025**. The degree consists of 123 credit points. My overall grade is **2.7 (satisfactory)**.

This notebook verifies the official documents (`urkunde.pdf` and `grades_masters.pdf`) against the structured metadata in `key_details.csv`. It also explains my course structure and grades.

```python
import pandas as pd
import os
from pypdf import PdfReader

# Define paths to the official documents and metadata
BASE_PATH = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
URKUNDE_PATH = os.path.join(BASE_PATH, 'urkunde.pdf')
GRADES_PATH = os.path.join(BASE_PATH, 'grades_masters.pdf')
KEY_DETAILS_CSV = os.path.join(BASE_PATH, 'data', 'key_details.csv')

def extract_text_from_pdf(pdf_path):
    """Extract full text from PDF for verification."""
    full_text = ""
    try:
        reader = PdfReader(pdf_path)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                full_text += page_text + "\n"
    except Exception as e:
        print(f"Error reading PDF: {e}")
    return full_text.lower()

# Extract text from both official documents
urkunde_text = extract_text_from_pdf(URKUNDE_PATH)
grades_text = extract_text_from_pdf(GRADES_PATH)
full_pdf_text = urkunde_text + grades_text

print("PDF Extraction Complete")
print("=" * 50)
print(f"Total characters extracted: {len(full_pdf_text)}")
print("First 600 characters for reference (urkunde.pdf):")
print(urkunde_text[:600])
```
